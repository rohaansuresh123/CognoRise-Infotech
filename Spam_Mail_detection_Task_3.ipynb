```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load your dataset. Modify the file path accordingly.
data = pd.read_csv('/content/spam.csv')

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data['Category'], data['Message'], test_size=0.2, random_state=42)

# Define models
models = [
    make_pipeline(TfidfVectorizer(), LogisticRegression()),
    make_pipeline(TfidfVectorizer(), RandomForestClassifier()),
    make_pipeline(TfidfVectorizer(), AdaBoostClassifier()),
    make_pipeline(TfidfVectorizer(), KNeighborsClassifier())
]

# Train and evaluate models
for i, model in enumerate(models):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)

    print(f"Model {i + 1}:\nAccuracy: {accuracy:.2f}\nClassification Report:\n{report}\n{'='*50}\n")
```
